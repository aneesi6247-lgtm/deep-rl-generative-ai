```python
import gymnasium as gym
import numpy as np
import random
```

```python
# sample environment
env = gym.make("CliffWalking-v1")
print(env.observation_space.n) # print the states
print(env.action_space.n) # action
init_state = env.reset() # 36 = 3*12+0
print(init_state)
```

# SARSA

```python
gamma = 0.99
alpha = 0.5
epsilon = 0.1 # means 10% exploration, 90% exploitation 
episodes = 500
```

```python
# Q_table >>> stores Q values
Q = np.zeros((48,4))
```

```python
# Policy -- episilon-greedy 

def epsilon_greedy(state): # for epsilon_greeedy approach we generate a random number
        if random.random() < epsilon: # this is the case of exploration
            return env.action_space.sample() # random action ==>> explore
        else:
            return np.argmax(Q[state]) # exploit
    
```

```python
for episode in range(episodes):

    env = gym.make("CliffWalking-v1")
    done = False # tell us when the episode has ended
    state,_ = env.reset()
    action= epsilon_greedy(state)

    total_reward= 0
    episode_length= 0

    while not done:
        next_state, reward, truncate, terminate, _ = env.step(action) # to take the first step and this return 5 parameters (_ shows additional info which we don't need)
        done = truncate or terminate 
        next_action = epsilon_greedy(next_state)

        # SARSA Update
        Q[state, action] += alpha * (reward + gamma * Q[next_state, next_action] - Q[state, action])
        state = next_state 
        action = next_action

        total_reward += reward
        episode_length +=1

    print(f"For the >>> {episode+1}/500, total reward {total_reward} and ep_leng {episode_length}")

                                    
        
```

```python
# what did our agent learn
env = gym.make("CliffWalking-v1", render_mode= "human")
state, _ =env.reset()
done = False
total_reward = 0
epi_length= 0
while not done:
    action = np.argmax(Q[state])
    next_state, reward, truncate, terminate, _ = env.step(action) # to take the first step and this return 5 parameters (_ shows additional info which we don't need)
    done = truncate or terminate
    state = next_state

    total_reward += reward
    epi_length +=1
print(f"For the >>> {episode+1}/500, total reward {total_reward} and ep_leng {epi_length}")
env.close()
```
